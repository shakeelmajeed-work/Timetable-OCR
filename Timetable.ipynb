Importing the required libraries

```python
import keras_ocr as ocr
import matplotlib.pyplot as plt
```

Setting up pipeline

```python
pipeline = ocr.pipeline.Pipeline()
```

Getting example images for training:

```python
images = [
    ocr.tools.read(img) for img in [
    'https://raw.githubusercontent.com/shakeelmajeed-work/Timetable-OCR/master/IMG_0429.jpg',
    'https://raw.githubusercontent.com/shakeelmajeed-work/Timetable-OCR/master/WhatsApp%20Image%202021-01-18%20at%2019.00.08.jpeg'
    ]
]
```

```python
%%html
<img src='https://raw.githubusercontent.com/shakeelmajeed-work/Timetable-OCR/master/IMG_0429.jpg'></img>
```

```python
prediction = pipeline.recognize(images)
print(prediction)
```

```python
# Plot the predictions
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction):
    ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
